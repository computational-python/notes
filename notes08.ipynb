## widgets

Consider plotting a function that depends on a paramter

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.sin(2*x))
```

```python
# define a function of the paramter

def sinplot(w):
    plt.plot(x, np.sin(w*x))

sinplot(1)
sinplot(2)
```

```python
#!uv pip install ipywidgets
import ipywidgets
```

```python
slider = ipywidgets.FloatSlider(min=1, max=2, value=2.0)
ipywidgets.interact(sinplot, w=slider)
```
