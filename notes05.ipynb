```python
class Counter:
    def __init__(self, size):
        print(f"__init__: initializing iterable with {size=}")
        self.size = size
        self.start = 0

    def __iter__(self):
        print(f"__iter__: starting up iteration with {self.size=}")
        return self

    def __next__(self):
        if self.start < self.size:
            self.start = self.start + 1
            return self.start
        raise StopIteration
```

```python
counter = Counter(3)
```

```python
def g(n):
    print('enter g with',n)
    yield n
    print('after yield')
```

```python
g(2)
```

```python
for i in g(2):
    print(i)
```

```python
g2 = g(2)
```

```python
next(g2)
```

```python
next(g2)
```

```python
def g(n):
    print('enter g with ',n)
    i=0
    while i < n:
        yield i
        print('after yield')
        i += 1
    print('after while')
```

```python
for i in g(5):
    print(i)
```

## the first n Fibonacci numbers

```python
def fib(n):
    a = 0
    b = 1
    count = 0
    while count < n:
        yield a
        a, b = b, a + b
        count += 1
```

```python
list(fib(1))
```

```python
list(fib(2))
```

```python
list(fib(3))
```

```python
list(fib(10))
```

```python
list(fib(20))
```

```python
import time

def tail(f):
    while True:
        line = f.readline()
        if not line:
            time.sleep(1)
            continue
        yield line
```

```python
for line in tail(open('123.txt')):
    print(line)
```

```python
def search_python(lines):
    for line in lines:
        if 'python' in line.lower():
            print(line)
```

```python
search_python(tail(open('123.txt')))
```

```python
## try-except block

try:
    search_python(tail(open('123.txt')))
except KeyboardInterrupt:
    print("Finshing")
```
