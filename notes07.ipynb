## Matrix multiplication

$$ C_{ij} = \sum_k A_{ik} B_{kj} $$

```python
import  time
n=512
a=numpy.ones((n,n))
b=numpy.ones((n,n))
c=numpy.zeros((n,n))
t1=time.time()
for i in range(n):
   for j in range(n):
       for k in range(n):
           c[i,j]+=a[i,k]*b[k,j]
t2=time.time()
print("Loop timing",t2-t1)
```

```python
a=numpy.ones((n,n))
b=numpy.ones((n,n))
c=numpy.zeros((n,n))
t1=time.time()
c = a @ b
t2=time.time()
print("Loop timing",t2-t1)
```

```python
%%file example.F90
INTEGER, PARAMETER :: N = 512 
REAL*8, DIMENSION(N,N) :: A, B, C
! Timing
INTEGER :: T1, T2, RATE
! Initialize
A = 1.0
B = 1.0
!
CALL SYSTEM_CLOCK(COUNT_RATE=RATE)
CALL SYSTEM_CLOCK(COUNT=T1)
C = MATMUL(A, B)
CALL SYSTEM_CLOCK(COUNT=T2)
PRINT '(A, F6.3)', 'MATMUL timing',  DBLE(T2-T1)/RATE
END
```
