# Concurrency

## Threading

```python
import threading
import time
```

```python
def work(seconds=1):
    print("Work...")
    time.sleep(seconds)
    print("Work done")
```

```python
t1 = time.time()
work(1)
work(2)
t2 = time.time()
print("Completed in ", t2-t1, "seconds")
```

```python
t1 = time.time()
thread1 = threading.Thread(target=work, args=[1])
thread2 = threading.Thread(target=work, args=[2])

thread1.start()
thread2.start()

thread1.join()
thread2.join()

t2 = time.time()
print("Completed in ", t2-t1, "seconds")
```

```python
t1 = time.time()

threads = []

for i in range(5):
    threads.append(threading.Thread(target=work, args=[i+1]))

for t in threads:
    t.start()
    

for t in threads:
    t.join()

t2 = time.time()
print("Completed in ", t2-t1, "seconds")
```

```python
import concurrent.futures
```

```python
t1 = time.time()
with concurrent.futures.ThreadPoolExecutor() as executor:
    executor.map(work, [1, 2, 3, 4, 5])

t2 = time.time()
print("Completed in ", t2-t1, "seconds")
```

## multiprocessing

```python
import multiprocessing
```

```python
t1 = time.time()

processes = []

for i in range(5):
    processes.append(multiprocessing.Process(target=work, args=[i+1]))

for p in processes:
    p.start()
    

for p in processes:
    p.join()

t2 = time.time()
print("Completed in ", t2-t1, "seconds")
```

```python
t1 = time.time()
with concurrent.futures.ProcessPoolExecutor() as executor:
    executor.map(work, [1, 2, 3, 4, 5])

t2 = time.time()
print("Completed in ", t2-t1, "seconds")
```
