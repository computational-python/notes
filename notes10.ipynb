# Concurrency

## Threading

```python
import threading
import time
```

```python
def work(seconds=1):
    print("Work...")
    time.sleep(seconds)
    print("Work done")
```

```python
t1 = time.time()
work(1)
work(2)
t2 = time.time()
print("Completed in ", t2-t1, "seconds")
```

```python
t1 = time.time()
thread1 = threading.Thread(target=work, args=[1])
thread2 = threading.Thread(target=work, args=[2])

thread1.start()
thread2.start()

thread1.join()
thread2.join()

t2 = time.time()
print("Completed in ", t2-t1, "seconds")
```

```python
t1 = time.time()

threads = []

for i in range(5):
    threads.append(threading.Thread(target=work, args=[i+1]))

for t in threads:
    t.start()
    

for t in threads:
    t.join()

t2 = time.time()
print("Completed in ", t2-t1, "seconds")
```

```python
import concurrent.futures
```

```python
t1 = time.time()
with concurrent.futures.ThreadPoolExecutor() as executor:
    executor.map(work, [1, 2, 3, 4, 5])

t2 = time.time()
print("Completed in ", t2-t1, "seconds")
```

## multiprocessing

```python
import multiprocessing
```

```python
t1 = time.time()

processes = []

for i in range(5):
    processes.append(multiprocessing.Process(target=work, args=[i+1]))

for p in processes:
    p.start()
    

for p in processes:
    p.join()

t2 = time.time()
print("Completed in ", t2-t1, "seconds")
```

```python
t1 = time.time()
with concurrent.futures.ProcessPoolExecutor() as executor:
    executor.map(work, [1, 2, 3, 4, 5])

t2 = time.time()
print("Completed in ", t2-t1, "seconds")
```

## asyncio

Demo cooking

```python
from time import sleep, time

def now(what):
    when = round(time() - start)
    print(when, what)


def prepare_salad():
    now("Cut salad")
    sleep(4)
    now("Rinse salad ")
    sleep(1)
    now("Salad done")


def prepare_potatoes():
    now("\tPeel potatoes")
    sleep(5)
    now("\tBoil potatoes")
    sleep(20)
    now("\tPotatoes done")


def prepare_steak():
    now("\t\tThaw steak")
    sleep(18)
    now("\t\tFry steak")
    sleep(2)
    now("\t\tSteak done")


def prepare_dinner():
    prepare_salad()
    prepare_potatoes()
    prepare_steak()
    now("Done")


start = time()
prepare_dinner()
# print(round(time() - start))
```

```python
%%file asyncdemo.py
from time import sleep, time
import asyncio

def now(what):
    when = round(time() - start)
    print(when, what)


async def prepare_salad():
    now("Cut salad")
    time.sleep(4)
    now("Rinse salad ")
    time.sleep(1)
    now("Salad done")


async def prepare_potatoes():
    now("\tPeel potatoes")
    time.sleep(5)
    now("\tBoil potatoes")
    await asyncio.sleep(20)
    now("\tPotatoes done")


async def prepare_steak():
    now("\t\tThaw steak")
    await asyncio.sleep(18)
    now("\t\tFry steak")
    await asyncio.sleep(2)
    now("\t\tSteak done")


async def prepare_dinner():
    task1 = asyncio.create_task(prepare_salad())
    task2 = asyncio.create_task(prepare_potatoes())
    task3 = asyncio.create_task(prepare_steak())

    await task1
    await task2
    await task3
    
    now("Done")


start = time()
asyncio.run(prepare_dinner())

# print(round(time() - start))
```

## MPI

```python
# MPI demo
"""
Parallel Hello World
"""
from mpi4py import MPI
import sys
size = MPI.COMM_WORLD.Get_size()
rank = MPI.COMM_WORLD.Get_rank()
name = MPI.Get_processor_name()
sys.stdout.write(
    "Hello, World! I am process %d of %d on %s.\n"
    % (rank, size, name))
```
