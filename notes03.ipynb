```python
%%file calculate.py
def add(a, b):
    """Return sum of two arguments
    >>> add(1, 1) #doctest: +SKIP
    2
    >>>
    """
    return a + b
def sub(a, b):
    """Return difference of two arguments
    >>> sub(1, 1) #doctest: +SKIP
    0
    """
    return a + b
```

```python
%%file test_calculate.py
import calculate

def test_add():
    calculated = calculate.add(1, 1)
    expected = 2
    assert calculated == expected, f'calculated: {calculated} != expected: {expected}'

def test_sub():
    calculated = calculate.sub(1, 1)
    expected = 0
    assert calculated == expected, f'calculated: {calculated} != expected: {expected}'
```

```python
!pytest test_calculate.py
```
