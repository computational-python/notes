```python
class Person:
    number = 0 #class attribute
    def __init__(self, given_name, surname):
        self.given_name = given_name # instance attributes
        self.surname = surname
        Person.number += 1

    def __str__(self):
        return f"{self.surname}, {self.given_name}"

    @staticmethod
    def report_number():
        print("Number of persons defined", Person.number)
```

```python
Person()
```

```python
p = Person('Olav', 'Vahtras')
```

```python
str(p)
```

```python
p.__str__()
```

```python
Person.__str__(p)
```

## inheritance

```python
class Employee(Person):
   def __init__(self, given_name, surname, salary):
       super().__init__(given_name, surname)
       #Person.__init__(self, given_name, surname)
       self.salary = salary
```

```python
emp = Employee('John', 'Doe', 100000)
```

```python
print(emp)
```

```python
emp.salary
```

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def action(self):
        print(self.name, 'is a', self.__class__.__name__)
        self.make_sound()
        self.chase()

    def make_sound(self):
        print("Don't know what", self.name, "sounds like")

    def chase(self):
        print("Don't know what", self.name, "chases")

class Dog(Animal):
    def make_sound(self):
        print(self.name, "says 'Bow-wow!'")

    def chase(self):
        print(self.name, "chases cat")

class Cat(Animal):
    def make_sound(self):
        print(self.name, "says 'Meow'!")

    def chase(self):
        print(self.name, "chases mouse")
```

```python
fido = Dog('Fido')
```

```python
fido.name
```

```python
fido.action()
```

```python
felix = Cat('Felix')
```

```python
felix.name
```

```python
felix.action()
```

```python
Animal('Unknown').action()
```
