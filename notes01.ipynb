## Iteration (Looping)

```python
print(1 * 7)
print(2 * 7)
print(3 * 7)
```

```python
i = 1
print(i * 7)
i = 2
print(i * 7)
i = 3
print(i * 7)
```

```python
for i in [1, 2, 3]:
    print(i * 7)
```

```python
for key in newdict:
    print(key)
    print(newdict[key])
```

```python
for item in newdict.items():
    print(item) # tuple
```

```python
for item in newdict.items():
    key, value = item # tuple unpacking
    print('key=',key)
    print('value=', value)
```

```python
for key, value in newdict.items():
    #key, value = item
    print('key=',key)
    print('value=', value)
```

```python
def print_table(n: int) -> None : # define function with a type hint
    """
    Print multiplication table for n

    input: n, integer > 0
    """
    print("Input value", n)
    for i in range(1, 11):
        #print(i, '*', n, '=', i * n)
        print(f'{i} * {n} = {i * n}')
```

```python
# call function
print_table(7)
```

```python
print(n) # n only defined inside function
```
